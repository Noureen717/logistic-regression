```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("Salary_dataset.csv")
```

```python
df.head()
```

```python
 df.info()
```

```python
df.shape
```

```python
x = df[['YearsExperience']]   # Independent variable
y= df['Salary']              # Dependent variable
```

```python
#Train-Test Split (it is a function) This function takes your complete data and shuffles it randomly before splitting it according to the specified parameters
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
)
#random_state=42: This is a "seed" used by the random number generator. The data is shuffled before splitting. Setting a specific integer value (like 42) ensures that the same random shuffle is applied every time you run the code
```

```python
#Create & Train Linear Regression Model

model = LinearRegression()
model.fit(x_train,y_train) # Trains a classifier or regressor.
```

```python
#Now Make Predictions ( test our trained model)
y_pred = model.predict(x_test)
y_pred
```

```python
print("Model Evaluation Summary")
print("-------------------------")
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
#VISUALISATION
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression: Salary vs Experience")
plt.show()
#Dots → Actual test data
#Line → Predicted regression line
#“Why doesn’t the regression line pass through all points?”
#Because linear regression finds the line that minimizes overall squared error rather than fitting each data point exactly.
```
